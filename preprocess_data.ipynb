```python
import pandas as pd
```

```python
df = pd.read_csv("/home/srinivasan/Skan/Models/kronos_curr/data/cisive_data.csv")
```

```python
df.head()
```

```python
act_counts = df['activity_discovered_name'].value_counts()
act_counts
```

```python
df2 = act_counts.reset_index()
```

```python
df2
```

```python
# Calculate the total sum of counts
total_counts = df2['count'].sum()

# Calculate the cumulative sum of counts
df2['cumulative_counts'] = df2['count'].cumsum()

# Calculate the cumulative percentage
df2['cumulative_percentage'] = df2['cumulative_counts'] / total_counts * 100

# Set a cumulative percentage threshold for filtering, e.g., filter out activities contributing less than 95% cumulatively
threshold = 95.0

# Filter out activities based on cumulative percentage
filtered_df = df2[df2['cumulative_percentage'] <= threshold]
```

```python
df2
```

```python
filtered_df
```

```python
df_filtered = df[df['activity_discovered_name'].isin(filtered_df['activity_discovered_name'])]
```

```python
ep_counts = df_filtered['event_path'].value_counts()
ep_counts
```

```python
df3 = ep_counts.reset_index()
```

```python
df3
```

```python
# Calculate the total sum of counts
total_counts = df3['count'].sum()

# Calculate the cumulative sum of counts
df3['cumulative_counts'] = df3['count'].cumsum()

# Calculate the cumulative percentage
df3['cumulative_percentage'] = df3['cumulative_counts'] / total_counts * 100

# Set a cumulative percentage threshold for filtering, e.g., filter out activities contributing less than 95% cumulatively
threshold = 95.0

# Filter out activities based on cumulative percentage
filtered_filtered_df = df3[df3['cumulative_percentage'] <= threshold]
```

```python
filtered_filtered_df
```

```python
df_filtered_filtered = df_filtered[df_filtered['event_path'].isin(filtered_filtered_df['event_path'])]
```

```python
df_filtered_filtered['active_url'] = df_filtered_filtered['url'].fillna("")
```

```python
df_filtered_filtered.drop(columns=['url'], axis=1, inplace=True)
```
